### Linear Regression

```python
import statsmodels.formula.api as sm
```

```python
water_2015
```

```python
formula1 = 'tp_totpop~ps_wfrto'
model1 = sm.ols(formula = formula1, data = water_2015)
fitted1 = model1.fit()
print(fitted1.summary())
```

```python


formula2 = 'water_scarcity_index ~ tp_totpop + ps_wfrto'
model2 = sm.ols(formula = formula2, data = water_scarcity)
fitted2 = model2.fit()
print(fitted2.summary())
```
