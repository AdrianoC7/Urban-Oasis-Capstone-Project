# us_cities_county processing #

```python
us_cities_county = pd.read_csv('uscities.csv')
us_cities_county = us_cities_county.drop(labels=['city_ascii','county_fips','population','density','source','military','incorporated','timezone','ranking','zips','id'],axis=1)
us_cities_county['lng']= us_cities_county['lng'].round(2)
us_cities_county['lat']= us_cities_county['lat'].round(2)
us_cities_county['lng']= us_cities_county['lng']*-1
```

```python
us_cities_county.head(500)
```

# us_cities_temp processing #

```python
us_cities_temp = pd.read_csv('glt_us_cities_long_lat.csv')
us_cities_temp.rename (columns={'Longitude':'lng','Latitude':'lat','City':'city'},inplace=True)
del us_cities_temp['Unnamed: 0']
us_cities_temp_pivot = us_cities_temp.pivot_table(index=['city','lat','lng'],columns='Year 2010',values='AverageTemperature')
us_cities_temp_pivot = us_cities_temp_pivot.reset_index()
us_cities_temp_pivot['lat'] = us_cities_temp_pivot['lat'].str.replace('N','')
us_cities_temp_pivot['lng'] = us_cities_temp_pivot['lng'].str.replace('W','')
us_cities_temp_pivot
```

# Joining the dataframes #

```python
cities_county_temp = pd.merge(us_cities_temp_pivot, us_cities_county, on=['city'],how='inner', suffixes=('_temp', '_county'))
cities_county_temp
```

```python
cities_county_temp['lat_temp'] = cities_county_temp['lat_temp'].astype(float)
cities_county_temp['lng_temp'] = cities_county_temp['lng_temp'].astype(float)
cities_county_temp['lat_county'] = cities_county_temp['lat_county'].astype(float)
cities_county_temp['lng_county'] = cities_county_temp['lng_county'].astype(float)
```

```python
cities_county_temp_filter = cities_county_temp[(cities_county_temp['lat_temp']-cities_county_temp['lat_county']).abs() <=.8]
cities_county_temp_filter = cities_county_temp_filter[(cities_county_temp_filter['lng_temp']-cities_county_temp_filter['lng_county']).abs() <=.8]
cities_county_temp_filter = cities_county_temp_filter.reset_index()
cities_county_temp_filter
```

# Cleaning the dataframe #

```python
del cities_county_temp_filter['index']
del cities_county_temp_filter['lat_county']
del cities_county_temp_filter['lng_county']
cities_county_temp_filter
```

```python
part1 = cities_county_temp_filter.loc[:, :'lng_temp']
part2 = cities_county_temp_filter.loc[:, '2010-01-01':'2010-12-01']
part3 = cities_county_temp_filter.loc[:, 'state_id':'county_name']
```

```python
cities_county_temp_final = pd.concat([part1,part3,part2],axis=1)
cities_county_temp_final
```

```python
cities_county_temp_final.to_csv('cities_county_temp_final.csv')
```
