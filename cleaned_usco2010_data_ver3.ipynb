### Chi square and ttests

```python
from scipy.stats import chi2_contingency
import pingouin
```

```python
water_df_copy = water_2010[["state","statefips","county","countyfips","fips","year","tp_totpop","ps_topop",
                            "ps_wgwto","ps_wswto","ps_wtotl","ir_wfrto", "ic_wfrto","to_wgwfr","to_wgwsa",
                            "to_wgwto","to_wswfr","to_wswsa","to_wswto","to_wfrto","to_wsato","to_wtotl"]].copy()
water_df_copy
```

```python
corr_df = water_df_copy.drop(columns=["statefips","year","countyfips","fips"]).copy()
corr_df 
```

```python
corr = corr_df.corr()
corr
```

```python
# Set the figure size
plt.figure(figsize=(20, 20))

# Create a heat map
sns.heatmap(corr, 
            xticklabels=corr.columns,
            yticklabels=corr.columns,
            annot=True, # show values in cells
            fmt=".2f", # specify the format of the values
            linewidths=.5, # space between cells
            cbar_kws={"shrink": .5}) # fix colorbar

# Saving the image
plt.savefig('water_correlation_2010.png')

# Show the plot
plt.show()
```
